# CREATING A MULTI-TABLE DATABASE DESIGN

###### Creating a connection

```python
import mysql.connector
import configparser
import pandas as pd
```

```python
config = configparser.ConfigParser()
config.read('My_SQL.ini')
try:
    mydb = mysql.connector.connect(
        host = config.get("My_SQL", "host"),
        port = config.get("My_SQL", "port"),
        user = config.get("My_SQL", "user"),
        password = config.get("My_SQL", "password"),
        database = config.get("My_SQL", "database"))
    cur = mydb.cursor(buffered=True)
    print("connected")
except:
    print("unable to connect")
```

##### Previewing the Parent Table

```python
df = pd.read_sql_query("SELECT * FROM kaggle_survey_2022_responses", mydb)
df.head()
```

##### Adding a primary key to the parent table

```python
cur.execute("""
ALTER TABLE kaggle_survey_2022_responses,
ADD COLUMN respondent_id INT AUTO_INCREMENT FIRST,
ADD PRIMARY KEY (respondent_id)""")
mydb.commit()
```

##### Viewing all columns

```python
cur.execute("""
SHOW COLUMNS FROM kaggle_survey_2022_responses
""")
columns = cur.fetchall()
for col in columns:
    print(col)
```

#### Table1 -countries

```python
#Creating table
cur.execute(""" 
CREATE TABLE countries 
(country_id INT PRIMARY KEY AUTO_INCREMENT, 
 country_name VARCHAR(100)
 )
""")
mydb.commit()
```

```python
#Inserting data
cur.execute("""
INSERT INTO countries(country_name)
SELECT country as c
FROM kaggle_survey_2022_responses
GROUP BY c
ORDER by c
""")
mydb.commit()
```

```python
#Viewing the table in python
cur.execute("SELECT * FROM countries")
results = cur.fetchall()
for x in results:
    print(x)
```

```python
#linking countries to parent table via FOREIGN KEY
cur.execute("""
ALTER TABLE kaggle_survey_2022_responses
ADD COLUMN country_id INT AFTER gender,
ADD CONSTRAINT fk_country
FOREIGN KEY (country_id)
REFERENCES countries(country_id)
""")
mydb.commit()
```

```python
#UPDATING the parent table using INNER JOIN
cur.execute("""
UPDATE kaggle_survey_2022_responses k
INNER JOIN countries c
ON c.country_name = k.country
SET k.country_id = c.country_id
""")
mydb.commit()
```

```python
#DROPING the country column from parent table
cur.execute("""
ALTER TABLE kaggle_survey_2022_responses
DROP COLUMN country
""")
mydb.commit()
```

```python
#Confirming the changes
df = pd.read_sql_query("SELECT * FROM kaggle_survey_2022_responses", mydb)
df.head()
```

```python
#Inserting data
cur.execute("""
INSERT IGNORE INTO respondents_data_science_activities(respondent_id, activity_id)
SELECT k.respondent_id, d.activity_id FROM kaggle_survey_2022_responses k, data_science_activities d
WHERE  
d.activity = k.activities_at_work1
OR
d.activity = k.activities_at_work2
OR
d.activity = k.activities_at_work3
OR
d.activity = k.activities_at_work4
OR
d.activity = k.activities_at_work5
OR
d.activity = k.activities_at_work6
OR
d.activity = k.activities_at_work7
OR
d.activity = k.activities_at_work8
""")
mydb.commit()
```
