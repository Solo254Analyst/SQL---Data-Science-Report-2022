# DATA PREPARATION

###### Importing the required libraries

```python
import configparser
import pymysql
pymysql.install_as_MySQLdb()
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

###### Creating a connection

```python
config = configparser.ConfigParser()
config.read('My_SQL.ini')
host = config.get("My_SQL", "host").replace('"', '')
port = config.get("My_SQL", "port").replace('"', '')
user = config.get("My_SQL", "user")
password = config.get("My_SQL", "password")
database = config.get("My_SQL", "database")
try:
    engine = create_engine(f'mysql://{user}:{password}@{host}:{port}/{database}')
    print("connected")
except:
    print("unable to connect")
```

###### Loading data (excel file) to dataframe

```python
df = pd.read_excel("kaggle_survey_2022_responses.xlsx")
df.shape
```
